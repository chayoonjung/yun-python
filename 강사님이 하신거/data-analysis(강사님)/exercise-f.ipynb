```python
import numpy as np
import analysis_helper as helper
```

```python
region = '양평제1동'
result = helper.pop_ratio_by_region(region)
if result is not None:
    
    import csv
    import numpy as np

    total = []
    with open('data-files/population.csv', encoding='utf-8') as f:
        reader = csv.reader(f)
        lst = list(reader)
        lst = lst[1:]
        for row in lst:
            if region in row[0]:
                total.append(100)
            else:
                result2 = np.array([ int(d.replace(',', '')) for d in row[3:] ])
                result2 = result2 / np.sum(result2)
                gap = result - result2
                # total.append(np.sum(np.power(gap, 2))) # 차이의 총량    
                total.append(np.sum(np.abs(gap))) # 차이의 총량    
    # print(np.array(total))
    # print(np.argmax(total), np.argmin(total),
    #      np.max(total), np.min(total))
    
    print(lst[np.argmin(total)])
    print(lst[np.argmax(total)])
    
    region1 = lst[np.argmax(total)] # my region
    region2 = lst[np.argmin(total)] # other region
    
    region1 = [ int(x.replace(',', '')) for x in region1[3:] ]
    region2 = [ int(x.replace(',', '')) for x in region2[3:] ]
    
    import matplotlib.pyplot as plt
    
    plt.plot(region1, label='region 1', c='red')
    plt.plot(region2, label='region 2', c='green')
    plt.show()
    
else:
    print("검색 실패")
```
