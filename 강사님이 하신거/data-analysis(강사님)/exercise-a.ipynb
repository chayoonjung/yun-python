```python
 # -3.14 ~ 3.14 사이의 256개로 균등 분할된 데이터 반환
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X) # X에 대한 cos값 반환
S = np.sin(X) # X에 대한 sin값 반환
plt.plot(X, C, label="cos")
plt.plot(X, S, label="sin")
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])
plt.legend(loc=2)
plt.show()
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1) # 2행 1열의 1번째 plot
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2) # 2행 1열의 2번째 plot
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

```python
# 매년 내 생일의 최고 기온 데이터를 리스트에 담으세요 (날짜 - 기온)
import csv
filtered_data = []
with open('data-files/seoul-temperature2.csv', encoding="utf-8") as f:
    data = csv.reader(f)
    next(data)
    month, day = '10', '01'
    for idx, row in enumerate(data):
        date = row[2] # yyyy-MM-dd
        date = date.split('-') # '1907-10-01' -> ['1907', '10', '01']
        if date[1] == month and date[2] == day: # 생일 비교
            filtered_data.append([row[2], row[4]])
```

```python
filtered_data[:3]
```

```python
# 추출한 데이터 중에서 
# 가장 높은 기온과 낮은 기온 및 평균 기온 계산 및 출력하세요

mx, mn, total = -100, 100, 0
mx_date = mn_date = ''
temperatures = [] # plot에서 사용할 데이터 (y축의 값으로 사용할 온도)
for row in filtered_data:
    # print(row)
    try:
        tmp = float(row[1]) # 온도를 숫자로 변환 : '17.1' -> 17.1
        total += tmp # 온도를 누적 합계 (평균 계산을 위해)
        temperatures.append(tmp) # 숫자 온도를 목록에 추가
        if mx < tmp:
            mx = tmp
            mx_date = row[0]
        if mn > tmp:
            mn = tmp
            mn_date = row[0]
    except:
        pass    
print(mx, mx_date, mn, mn_date, total / len(filtered_data)) 
    
```

```python
import matplotlib.pyplot as plt

avg = total / len(filtered_data)
plt.plot(temperatures)
plt.plot([avg] * len(temperatures), 'r--')
plt.ylim(0, mx + 6)
plt.show()
```

```python
import csv
filtered_data = []
cnt = total = 0 # 정상 데이터의 정보 저장하는 변수
with open('data-files/seoul-temperature2.csv', encoding="utf-8") as f:
    data = csv.reader(f)
    next(data)
    for row in data:
        if row[4] != '': # 결측치가 아닌 경우
            filtered_data.append( float(row[4]) )
            total += float(row[4])
            cnt += 1
        else:
            filtered_data.append( -1000 )
    
# 전체 데이터를 반복하면서 결측치를 평균으로 대체
mean = total / cnt # 정상데이터의 평균
for idx in range(len(filtered_data)) :
    if filtered_data[idx] == -1000:
        filtered_data[idx] = mean
    
```

```python
print(mean, sum(filtered_data) / len(filtered_data))
```

```python
import matplotlib.pyplot as plt

plt.hist(filtered_data, bins=100, color='red')
plt.show()
```
