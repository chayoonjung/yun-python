```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
cctv = pd.read_csv('data-files/a.cctv-in-seoul.csv')
```

```python
cctv.head(3)
```

```python
print( list(cctv.columns) )
print( list(cctv.index) )
```

```python
# cctv.columns = list('abcdef')
cctv.columns = \
    ['기관명', '소계', '2013년도 이전', '2014년', '2015년', '2016년']
cctv.head(2)
```

```python
cctv.rename(columns={ "기관명": "구별" },
            inplace=True)
cctv.head(2)
```

```python
cctv.sort_values(by="소계", ascending=False).head(3)
```

```python
a = cctv.iloc[:, 3:6].sum(axis=1) # 최근 3년 합계
b = cctv.iloc[:, 2] # 4년 이전에 설치된 CCTV 개수
cctv['최근증가율'] = a / b * 100
cctv.sort_values(by='최근증가율', ascending=False)
```

```python
# pd.read_excel('data-files/a.population-in-seoul.xls')
pop_in_seoul = \
    pd.read_excel('data-files/a.population-in-seoul.xls',
                  header=2,
                  usecols='B, D, G, J, N')
```

```python
pop_in_seoul.head(3)
pop_in_seoul.tail(3)
pop_in_seoul = pop_in_seoul.iloc[1:-1]
pop_in_seoul
```

```python
pop_in_seoul2 = pop_in_seoul.copy()
```

```python
# pop_in_seoul.columns = \
#     ['구별', '인구수', '한국인', '외국인', '고령자']
# pop_in_seoul.head(3)
names = ['구별', '인구수', '한국인', '외국인', '고령자']
pop_in_seoul.rename(
    columns={ p:n for p, n in zip(pop_in_seoul.columns, names) },
    inplace=True)
pop_in_seoul.head()
```

```python
pop_in_seoul['구별'].unique()
pop_in_seoul['구별'].value_counts()
```

```python
pop_in_seoul['외국인비율'] = \
    pop_in_seoul['외국인'] / pop_in_seoul['인구수'] * 100
pop_in_seoul['고령자비율'] = \
    pop_in_seoul['고령자'] / pop_in_seoul['인구수'] * 100
# pop_in_seoul.sort_values(by="외국인비율", ascending=False)
# pop_in_seoul.sort_values(by="고령자비율", ascending=False)
pop_in_seoul.sort_values(by="인구수", ascending=False)
```

```python
pop_in_seoul.head(3)
```

```python
cctv.head(3)
```

```python
print( cctv["구별"].sort_values() )
print( pop_in_seoul["구별"].sort_values() )
```

```python
# pandas.merge : 두 테이블을 병합하는 함수 (같은 이름의 컬럼 기준)
# merged_data = pd.merge(cctv, pop_in_seoul)
merged_data = pd.merge(cctv, pop_in_seoul, on='구별')
```

```python
merged_data.head(3)
# merged_data.shape
```

```python
# merged_data.drop( \
#     ['2013년도 이전', '2014년', '2015년', '2016년'], 
#     axis=1,
#     inplace=True)
merged_data = merged_data.iloc[:, [0, 1, *range(6, 13)]]
```

```python
merged_data.head(3)
```

```python
# set_index : 지정된 컬럼을 인덱스로 사용
merged_data.set_index('구별', inplace=True)
```

```python
merged_data.head()
```

```python
# reset_index : 인덱스 컬럼을 일반 컬럼으로 변경
merged_data.reset_index().head(3)
```

```python
merged_data.sort_values('소계', ascending=False).head()
```

```python
merged_data.sort_values('인구수', ascending=False).head()
```

```python
try:
    del merged_data['인구당비율']
except:
    pass
try:
    del merged_data['CCTV비율']
except:
    pass

merged_data['CCTV당인구'] = \
    merged_data['인구수'] / merged_data['소계']
```

```python
merged_data.sort_values(by='CCTV당인구', ascending=True).head()
```
