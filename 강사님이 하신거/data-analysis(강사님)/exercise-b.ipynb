```python
import csv

population = { "female": [], "male": [] }

with open('data-files/population2.csv', encoding="utf-8") as f:
    reader = csv.reader(f)
    next(reader) # skip header row
    for row in reader:
        if '양평제1동' in row[0]: 
            for cnt in row[3:104]:
                population['male'].append( int(cnt.replace(',', '')) )
            for cnt in row[106:]:
                population['female'].append( int(cnt.replace(',', '')) )
            break
```

```python
import csv

def find_data_by_region(region_name):
    population = { "female": None, "male": None }

    with open('data-files/population2.csv', encoding="utf-8") as f:
        reader = csv.reader(f)
        next(reader) # skip header row
        for row in reader:
            if region_name in row[0]: 
                population['male'] = \
                    [int(cnt.replace(',', '')) for cnt in row[3:104]]
                population['female'] = \
                    [int(cnt.replace(',', '')) for cnt in row[106:]]
                break
                
    return population
```

```python
import matplotlib.pyplot as plt

population = find_data_by_region('발산제1동')

plt.figure(figsize=(15, 10))
plt.plot(population['female'], label='female')
plt.plot(population['male'], label='male')
plt.legend(loc=2)
plt.show()
```
