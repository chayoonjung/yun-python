```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
tips = sns.load_dataset("tips")
tips.info()
tips.head()
```

```python
pd.pivot_table(tips, index=["day"], values=["tip"])
pd.pivot_table(tips, index=["smoker"], values=["tip"])
pd.pivot_table(tips, index=["sex"], values=["tip"])
pd.pivot_table(tips, index=["time"], values=["tip"])
pd.pivot_table(tips, index=["time"], values=["total_bill"])
tips.groupby("time")["total_bill"].mean()
tips.groupby('size')["tip"].mean()
```

```python
tips[["tip", "size"]].corr()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
# sns.relplot(x="total_bill", y="tip", 
#             hue="smoker", style="time", 
#             data=tips)            
sns.scatterplot(x="total_bill", y="tip", 
                hue="smoker", style="time", 
                data=tips)

plt.show()
```

```python
# sns.relplot(x="total_bill", y="tip", hue="size", data=tips)
sns.relplot(x="total_bill", y="tip", size="size", data=tips)
plt.show()
```

```python
my_map = \
    folium.Map(location=[37.566688, 126.978419], 
               #tiles="Stamen Toner",
               zoom_start=11)

my_map.choropleth(geo_data=geo_simple,
                  data=crime_in_seoul[""],                  
                  fill_color="PuRd",
                  key_on='feature.id')
my_map
```

```python
my_map = \
    folium.Map(location=[37.566688, 126.978419], zoom_start=11)

for lat, lng in lat_lng:
    folium.Marker([lat, lng]).add_to(my_map)

my_map
```

```python
cis = \
    pd.read_csv('data-files/b.crime-in-seoul.csv', 
                encoding="utf-8", thousands=",")
```

```python
cis[''] = cis.iloc[:, 1:10:2].sum(axis=1)
cis
```

```python
my_map = \
    folium.Map(location=[37.566688, 126.978419], zoom_start=11)

for idx, loc in enumerate(lat_lng):    
    folium.CircleMarker(loc,
                        radius=float(cis.iloc[idx,-1])/100,
                        color="red", fill_color="red",
                        fill=True).add_to(my_map)

my_map
```
