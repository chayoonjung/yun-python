```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
store = pd.read_csv('sg-data/상가업소정보_서울.csv', encoding='utf-8')
store.shape
```

```python
# store.head()   # tail은 아래에서부터
```

```python
store.keys()
```

```python
view_columns = ['상호명','상권업종대분류명','상권업종중분류명','상권업종소분류명',
               '시도명','시군구명','행정동명','법정동명','지번주소','경도', '위도']
store_t=store[view_columns]
```

```python
store_t.head()
```

```python
# 결측치 확인
store_t.isnull().sum()

import missingno as msno

# msno.matrix(store)
```

```python
store_t = store_t.dropna()
```

```python
store_t['시도명'].unique()
store_t['시군구명'].unique()
store_t['시군구명'].value_counts()
store_t['상권업종대분류명'].unique()
```

```python
plt.rc('font', family="Malgun Gothic")
plt.figure(figsize=(15,4))
sns.countplot(data=store_t, x="시군구명", order=store_t['시군구명'].value_counts().index)
# plt.show()
```

```python
store_t.plot.scatter(x="경도", y="위도",figsize=(8,8), grid = True )
```

```python
plt.figure(figsize=(11,9))
sns.scatterplot(data=store_t, x="경도", y="위도", hue="시군구명")     #hue=색상
```
