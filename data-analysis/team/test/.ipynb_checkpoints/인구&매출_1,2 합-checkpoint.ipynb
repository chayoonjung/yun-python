```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_01 = pd.read_csv('sg-data/origin/01.상권분석서비스(상권-추정유동인구).csv', encoding='EUC-KR')
# df_02 = pd.read_csv('sg-data/origin/02.상권분석서비스(상권-집객시설).csv', encoding='EUC-KR')
df_03 = pd.read_csv('sg-data/origin/03.상권분석서비스(상권배후지-추정유동인구).csv', encoding='EUC-KR')
# df_04 = pd.read_csv('sg-data/origin/04.상권분석서비스(상권배후지-집객시설).csv', encoding='EUC-KR')
df_05 = pd.read_csv('sg-data/origin/05.상권분석서비스(상권배후지-상주인구).csv', encoding='EUC-KR')
df_06 = pd.read_csv('sg-data/origin/06.상권분석서비스(상권_상주인구).csv', encoding='EUC-KR')
# df_07 = pd.read_csv('sg-data/origin/07.상권분석서비스(상권배후지-아파트).csv', encoding='EUC-KR')
# df_08 = pd.read_csv('sg-data/origin/08.상권분석서비스(상권배후지-점포).csv', encoding='EUC-KR')
# df_09 = pd.read_csv('sg-data/origin/09.상권분석서비스(상권-상권변화지표).csv', encoding='EUC-KR')
# df_10 = pd.read_csv('sg-data/origin/10.상권분석서비스(행정동별 상권변화지표).csv', encoding='EUC-KR')
df_11 = pd.read_csv('sg-data/origin/11.상권분석서비스(상권-추정매출)_2019.csv', encoding='EUC-KR')
```

```python
#2019년도만 뽑기
df_01_2019 = df_01[ df_01['기준 년코드'] == 2019 ]  #상권-유동인구
df_03_2019 = df_03[ df_03['기준_년_코드'] == 2019 ] #배후지-유동인구
df_05_2019 = df_05[ df_05['기준_년_코드'] == 2019 ] #배후지-상주인구
df_06_2019 = df_06[ df_06['기준_년_코드'] == 2019 ] #상권-상주인구
df_11_2019 = df_11[ df_11['기준_년_코드'] == 2019 ] #상권-추정매출
```

```python
# 컬럼 name 변경
df_03_2019=df_03_2019.rename(columns={'기준_분기_코드                               ':"기준_분기_코드"})  # 컬럼 rename 설정
df_01_2019=df_01_2019.rename(columns={" 상권_구분_코드":'상권_구분_코드', " 상권_구분_코드_명":"상권_구분_코드_명"})
df_06_2019=df_06_2019.rename(columns={'상권 코드':"상권_코드"})  # 컬럼 rename 설정
```

```python
# 각 파일별 분기 확인
df_01_2019['기준_분기_코드'].unique()    #1,2
# df_03_2019.keys()
df_03_2019['기준_분기_코드'].unique()     #1,2
df_05_2019['기준_분기_코드'].unique()  #1,2,3,4
df_06_2019['기준_분기_코드'].unique()  #1,2,3,4
df_11_2019['기준_분기_코드'].unique()  #1,2,3,4
```

```python
#2019 1, 2 분기만 뽑기
df_01_2019_1 = df_01_2019[ df_01_2019['기준_분기_코드'] == 1 ]
df_01_2019_2 = df_01_2019[ df_01_2019['기준_분기_코드'] == 2]

df_03_2019_1 = df_03_2019[ df_03_2019['기준_분기_코드'] == 1 ]
df_03_2019_2 = df_03_2019[ df_03_2019['기준_분기_코드'] == 2]

df_05_2019_1 = df_05_2019[ df_05_2019['기준_분기_코드'] == 1 ]
df_05_2019_2 = df_05_2019[ df_05_2019['기준_분기_코드'] == 2]
df_05_2019 = df_05_2019_1.append(df_05_2019_2)

df_06_2019_1 = df_06_2019[ df_06_2019['기준_분기_코드'] == 1 ]
df_06_2019_2 = df_06_2019[ df_06_2019['기준_분기_코드'] == 2]
df_06_2019 = df_06_2019_1.append(df_06_2019_2)

df_11_2019_1 = df_11_2019[ df_11_2019['기준_분기_코드'] == 1 ]
df_11_2019_2 = df_11_2019[ df_11_2019['기준_분기_코드'] == 2]
df_11_2019 = df_11_2019_1.append(df_11_2019_2)
```
