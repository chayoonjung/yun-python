```python
x1 = np.linspace(0.0, 5.0) 
x2 = np.linspace(0.0, 2.0) 
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1) 
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1) # 2행 1열 1번째 plot
plt.plot(x1, y1, 'yo-') 
plt.title('A tale of 2 subplots') 
plt.ylabel('Damped oscillation') 


ax2 = plt.subplot(2, 1, 2) # 2행 1열 2번째 plot
plt.plot(x2, y2, 'r.-') 
plt.xlabel('time (s)') 
plt.ylabel('Undamped') 

plt.show()
```

```python
#일자별 온도 데이터 파일 사용
#매년 내 생일의 기온 데이터를 리스트에 담으시오.
#가장 높은기온, 낮은기온 및 평균기온 계산하여 출력.
#기온 데이터의 추세를 시각화해서 표현(line plot)

import csv
filtered_data = []
with open('data-files/seoul-temperature2.csv',
           encoding='utf-8') as f:
    data = csv.reader(f) 
    next(data)
    month, day = '09', '30'
    for idx, row in enumerate(data):    #enumerate(LIST) ->순번, data
        date = row[2] # yyyy-mm-dd
        date = date.split('-') # '1907 - 10 - 01' ->[1907, 10, 1]
        if date[1] == month and date[2] == day:  # 생일비교
            filtered_data.append([row[2],row[4]])
            
```

```python
#가장 높은기온, 낮은기온 및 평균기온 계산하여 출력.
mx, mn, total = -100, 100, 0
mx_date = mn_date = ''
temperatures = []    # plot에서 사용할 데이터. y캆 으로 사용할 온도 데이터
for row in filtered_data:
    
    try:
        tmp = float(row[1]) #온도를 숫자로 변환 '17.1'-> 17.1
        total += tmp # 온도 누적 합계 (평균 계산을 위해서)
        temperatures.append(tmp) #숫자 온도를 목록에 추가
        
        if mx < tmp:
            mx = tmp
            mx_date = row[0]    # filtered_data 에서 나온 ['1908-09-30', '21.8'] 순으로 나오므로 row[0]을 넣어야 함.
        if mn > tmp:
            mn = tmp
            mn_date = row[0]
    except :
        pass
        
print([mx, mx_date, mn, mn_date], '평균:', total / len(filtered_data)) 
```

```python
import matplotlib.pyplot as plt

avg = total / len(filtered_data)
plt.plot(temperatures)
plt.plot([avg]* len(temperatures),'r--')
plt.ylim(0, mx+10)
plt.show()
```

```python
#일자별 온도 데이터 파일 사용
#매년 내 생일의 기온 데이터를 리스트에 담으시오.
import csv

filtered_data = []
with open('data-files/seoul-temperature2.csv',
           encoding='utf-8') as f:
    data = csv.reader(f) 
    next(data)   #제목 줄 빼는 것
    month, day = '09', '30'
    filtered_data = [ [row[2],row[4]] for row in data \
                                      if row[4] != '' and            #missing value 제거
                                         row[2].split('-')[1] == month and
                                         row[2].split('-')[2] == day]
    #내포 : [expression for data in list_data if condition]
```

```python
#가장 높은기온, 낮은기온 및 평균기온 계산하여 출력.
mx, mn, total = -100, 100, 0
mx_date = mn_date = ''
temperatures = []    # plot에서 사용할 데이터. y캆 으로 사용할 온도 데이터
for row in filtered_data:
    
    try:
        tmp = float(row[1]) #온도를 숫자로 변환 '17.1'-> 17.1
        total += tmp # 온도 누적 합계 (평균 계산을 위해서)
        temperatures.append(tmp) #숫자 온도를 목록에 추가
        
        if mx < tmp:
            mx = tmp
            mx_date = row[0]    # filtered_data 에서 나온 ['1908-09-30', '21.8'] 순으로 나오므로 row[0]을 넣어야 함.
        if mn > tmp:
            mn = tmp
            mn_date = row[0]
    except :
        pass
        
print([mx, mx_date, mn, mn_date], '평균:', total / len(filtered_data)) 
```

```python
#일자별 온도 데이터 파일 사용
#매년 내 생일의 기온 데이터를 리스트에 담으시오.
#가장 높은기온, 낮은기온 및 평균기온 계산하여 출력.
#기온 데이터의 추세를 시각화해서 표현(line plot)

import csv
filtered_data = []
cnt = total = 0   # 정상 데이터의 정보 저장하는 변수
with open('data-files/seoul-temperature2.csv',
           encoding='utf-8') as f:
    data = csv.reader(f) 
    next(data)
    for row in data:
        if row[4] != '':      # 결측치가 아닌경우
            filtered_data.append(float(row[4]))
            total += float(row[4])   #결측치가 아닌 경우에만 합을 구하는 것.
            cnt += 1
        else:
            filtered_data.append( -1000 )         # -1000은 절대 나올수 없는 값을 임의로 작성함. '' 또는 'nan'을 작성해도 무관하다. 절대 나올수 없는 데이터를 작성.

#전체 데이터를 반복하면서 결측치를 평균으로 대체

mean = total / cnt  # 정상 데이터의 평균
for idx in range(len(filtered_data)): 
    if filtered_data[idx] == -1000:
        filtered_data[idx] = mean
      
```
