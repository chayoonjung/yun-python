```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
population_df = pd.read_csv('data-files/population.csv', 
                             encoding='utf-8', 
                             index_col=0)

population_df.head(3)
```

```python
columns_to_delete = population_df.columns[1]
columns_to_delete
```

```python
population_df.drop(population_df.columns[1], axis = 1)
population_df.drop(columns_to_delete, axis = 1, inplace=True) #drop 돌렸는데 또 돌리면 오류 난다!
```

```python
population_df.head(3)   
```

```python
remove_row = ~population_df.index.str.contains("000000")
population_df2 = population_df.loc[remove_row]
print(population_df.index.shape)
print(population_df2.index.shape)
```

```python
mask = population_df.index.str.contains('화곡제1동')
my_region = population_df[mask]
my_region
```

```python
print(my_region.values.shape)
my_region.values[0,1:]
```
