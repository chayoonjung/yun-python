```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
import scipy.stats
```

```python
np.random.seed(42)

N = 10      # 횟수 10번
mu_0 = 0.8   # 데이터 뽑기에 이용하는 평균값

x = sp.stats.bernoulli(mu_0).rvs(N) # x = 표본
print('승-패 : ' , x)

n = np.count_nonzero(x)  # 배열에서 0이 아닌 값의 개수를 계산
print('승리 횟수: [{0}] / 승률 : [{1}] '.format(n, n/N)) 


p_value = sp.stats.binom_test(n, N)  
#모집단 가설에서 표본값 나올 확률 
#귀무가설은 일반적으로 0.5로 설정되어 있다.
p_value
```

```python
# 카이 제곱 검정 (범주형 확률 분포에서 사용,  독립 또는 상관관계 존재 여부도 검증)
obs = np.array([[5, 25], [10, 20]])
print(obs)


# chi2_contingency() : 명령의 결과 튜플로 반환되며 chi2 값이 검정통계량, p_value 값이 유의확률이다.
chi2, p_value, dof , expected = sp.stats.chi2_contingency(obs)
print(chi2, dof, p_value)  #dof : 자유도

```

```python
obs = np.array([[4, 16, 20], [23, 18, 19]])
print(obs)

chi2, p_value, dof , expected = sp.stats.chi2_contingency(obs)
print(chi2, dof, p_value)
```

```python
# T분포 검정 (연속형데어터, 정규분포 표본 평균에 대한 기댓값을 검정)
np.random.seed(0) 

# N = 10 #데이터 10건
# mu_0 = 0 #평균값

# x = sp.stats.norm(mu_0).rvs(N) 
# sns.distplot(x, kde=False, fit=sp.stats.norm) 
# plt.show()

x = np.random.randint(60, 100, 10) #60에서 100 사이 10개 데이터 뽑기

#표본x가 평균 0인 모집단에서 추출되었는지 검정
print(x.mean()) #표본
# sp.stats.ttest_1samp(x, popmean=0)   #popmean=0 : 모집단평균 = 0 (귀무가설)
sp.stats.ttest_1samp(x, popmean=73) #popmean=73 : 모집단평균 = 73 (귀무가설)
# 모집단평균 = 73 에서 p_value 값이 유의확률이 0.527 이 나옴. => 우연히 50% 확률 만큼 나온다는 의미 /표본에서 우연히 그만큼 나왔다는 의미 
```

```python
np.random.seed(0) 

x1 = np.random.randint(70, 97, 20) # A학원 학생들의 수학성적
x2 = np.random.randint(65, 90, 20) # B학원 학생들의 수학성적

print(x1.mean(), x2.mean())

sp.stats.ttest_ind(x1, x2, equal_var=False)
# statistic : 데이터의 합을 데이터 포인트 수로 나눈 값, 데이터의 중심 위치, 통계
# ttest : 두 그룹은 비교하는 데이터 / 정규분포 이루고 있는 연속형 데이터인 경우 사용
# 모집단은 같은데 다르게 나올 확률 12% , 5%기준으로 보고 우연인지 필연인지 비교 (5% 넘었으므로 우연)
# 5% 기준으로 아래면 필연, 위면 우연
# equal_var=True : 표준편차 비슷하다.(분산도 고려)
```

```python
# 추정 및 검정 – 대응 이표본 평균
# 값의 전과 후 평균 비교

N = 5 
mu_1 = 0 
mu_2 = 0.4 
np.random.seed(1) 
x1 = sp.stats.norm(mu_1).rvs(N) 
x2 = x1 + sp.stats.norm(mu_2, 0.1).rvs(N)

print(x1)
print(x2)
print(x1.mean(), x2.mean()) 
print(sp.stats.ttest_ind(x1, x2, equal_var=False)) #pvalue 0.6 미미한 차이.
print(sp.stats.ttest_rel(x1, x2)) # 0.004 -> 차이가 크다. 
```
