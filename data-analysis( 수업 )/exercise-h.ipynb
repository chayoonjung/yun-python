```python
#  위 작업은 탐색 연습
#  아래는 일괄 처리 구현
```

```python
population_df = pd.read_csv('data-files/population.csv', 
                             encoding='utf-8', 
                             index_col=0)
population = population_df.copy()
```

```python
population = population.iloc[:, 2:] 
population.head()
```

```python
population['인구수합계'] = population.sum(axis=1) #그냥 sum() 하면 행으로 더해진다. 
population.head(2)
```

```python
# population["인구수합계"]
```

```python
population_ratio = population.div(population["인구수합계"],axis=0)
population_ratio.head(2)
```

```python
my_region = population_ratio[population_ratio.index.str.contains('화곡제1동')]
```

```python
my_region
```

```python
my_region.iloc[0]
```

```python
gap = population_ratio.sub(my_region.iloc[0], axis=1)
# gap = np.power(gap,2)
gap = np.abs(gap) # 절대값.
summed_list = gap.sum(axis=1)
top_5_similar_regions = summed_list.sort_values(ascending=True)[:5]    # 차이가 적울수록 비슷한 동네
top_5_similar_regions
```

```python
plt.rc('font',family = 'Malgun Gothic')
d = population_ratio.loc[top_5_similar_regions.index]
d = d.iloc[:, :-1]
# plt.figure(figsize=(10,10)) ->overwrite로 안된것임. 아래처럼 작성해야 된다.
d.T.plot(figsize=(10,7)) #데이터 프레임을 그래프로 출력 # T를 넣어야 행과 열이 바뀐다.
plt.show()
```
