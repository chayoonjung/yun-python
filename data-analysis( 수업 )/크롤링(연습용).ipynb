```python
import pandas as pd

df = pd.read_html('https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table',
                  index_col=0)
df[1]
```

```python
winter = df[1].iloc[:-1,5:10]
```

```python
winter.columns = ['games', 'gold', 'silver', 'bronze', 'total']
winter.head()
```

```python
print(winter.sort_values('gold', ascending = False))
```

```python
# winter["gold"].sum() , winter["gold"].mean() , winter['gold'].max(), winter['gold'].min()
winter['total'].sum(), winter["total"].mean() , winter['total'].max(), winter['total'].min()
```
