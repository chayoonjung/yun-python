```python
# 특정 지역의 성별, 나이별 인구수 데이터를 정수 리스트로 만드시오.
# line plot을 사용하여 성별로 나이별 인구수 추이를 확인하세요.
# bar plot (bar()함수)을 사용해서 성별로 나이별 인구수 분포 확인하시오.

import csv

male = []  # m: 남성, f: 여성 으로 변수 만들기
female = []

with open('data-files/population2.csv', encoding='utf-8') as f:
    data = csv.reader(f)
    next(data)
    # for row in data
    for row in data:
        if '화곡제1동' in row[0]:
            for i in row[3:104]:
                male.append(int(i.replace(',','')))
            for i in row[106:]:
                female.append(int(i.replace(',','')))
print(female)
print(male)
```

```python
import matplotlib.pyplot as plt

plt.plot(male, color = 'blue') 
plt.plot(female, color = 'red') 
plt.show()
```

```python
import csv

population = {"female": [], "male": []}   #None로 했을경우 dic / [] ->리스트 만들기

with open('data-files/population2.csv', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)   # skip header row  (이름부분 skip 하고 다음행부터 읽는다.)
    for row in reader:
        if '화곡제1동' in row[0]:    # 특정 지역 설정
            for cnt in row[3:104]:
                population['male'].append(int(cnt.replace(',','')))  
            for cnt in row[106:]:
                population['female'].append(int(cnt.replace(',','')))
            break
```

```python
import csv

def find_data_by_region(region_name):
    population = {"female": None, "male": None}   #None로 했을경우 dic / [] ->리스트 만들기

    with open('data-files/population2.csv', encoding='utf-8') as f:
        reader = csv.reader(f)
        next(reader)   # skip header row  (이름부분 skip 하고 다음행부터 읽는다.)
        for row in reader:
            if region_name in row[0]: 
                population['male'] = [int(cnt.replace(',','')) for cnt in row[3:104]]   
                population['female'] = [int(cnt.replace(',','')) for cnt in row[106:]]
                break
    return population
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))  # 그래프 사이즈 조정
population = find_data_by_region('발산제1동')
plt.plot(population['female'], label ='female')
plt.plot(population['male'], label ='male')
plt.legend(loc=2)   # 범례 설정
plt.show()
```
