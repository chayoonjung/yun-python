```python
import numpy as np
import analysis_helper as helper     # helper 함수를 가지고 와서 이용.(vscode에서 만들어서 이용함.)
```

```python
region = '화곡제1동'
result = helper.pop_ratio_by_region(region)
if result is not None:
    
    import csv
    import numpy as np

    total =[]
    with open('data-files/population.csv', encoding = 'utf-8') as f:
        reader = csv.reader(f)
        lst = list(reader)   #reader을 리스트로.
        lst = lst[1:]
        for row in lst:
            if region in row[0]:
                total.append(100)   # 큰 값을 넣어 큰차이 만들어 본다.
            else:
                result2 = np.array([ int(d.replace(',','')) for d in row[3:] ])  
                result2 = result2 / np.sum(result2)
                gap = result - result2
#                 total.append(np.sum(np.power(gap, 2)))  #power : 제곱  / 차이의 총량 
                total.append(np.sum(np.abs(gap)))         #abs => 절대값 
#     print(np.array(total))
#     print(np.argmax(total), np.argmin(total),              #argmax  , argmin  => 어느 위치의 가장 큰 데이터 ,작은데이터 있는지위치를 나타냄.  
#           np.max(total), np.min(total)) 
    
    print(lst[np.argmin(total)])    # total 제일 비슷한 지역 출력
    print(lst[np.argmax(total)])
    
    region1 = lst[np.argmax(total)]   # my region
    region2 = lst[np.argmin(total)]   # 비슷한 다른 지역(other region)
    
    region1 = [int(x.replace(',',"")) for x in region1[3:] ]
    region2 = [int(x.replace(',',"")) for x in region2[3:] ]
    
    import matplotlib.pyplot as plt
    
    plt.plot(region1, label = 'region1', c='red')
    plt.plot(region2, label = 'region2', c='blue')
    plt.show()       ##인구수가 아닌 비율을 보기 때문에 모양이 비슷한걸 찾은 것임.

else:
    print("검색실패")  #삼성1동 검색할때 시각화 보면 인구차이가 심해서 이상해 보임. region2 = [int(x.replace(',',""))/100 for x in region2[3:] ] 나누기 100을 해주면 볼수 있다.
```
